# Mean, Median, Mode, Variance and Standard Deviation
- <b> Mean</b> :- The sample mean is the average and is computed as the sum of all the observed outcomes  from the sample divided by the total number of events.  We use x as the symbol for the sample mean.
- <b> Median </b> :- In a sorted, ascending or descending, list of numbers, the median is the middle number and may be more representative of that data set than the average.
- <b> Mode </b> :- The mode is the value that most frequently appears in a data value set.
- <b>Standard Deviation </b> :- A calculation of the amount of variance or dispersion of a set of values is the standard deviation.
- <b> Variance </b> :- The expectation of the square deviation of a random variable from its mean is variance.

# Normal Distribution
- Normal distribution is also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.
- Features of Normal Distribution <br>
1. In normal distributation The mean, mode and median are all equal. 
    The curve is symmetric at the center (i.e. around the mean, μ).<br>
    Exactly half of the values are to the left of center and exactly half the values are to the right.<br>
    The total area under the curve is 1.<br>
    The skewness for Normal distribution is 0.<br>
    The kurtosis for Normal distribution is 3.<br>
2. Empirical rule for the normal distribution<br>
    
    68% of data lies in the one standard deviation away from the mean.<br>
    95% of data lies in the two standard deviation away from the mean.<br>
    99.7% of data lies in the three standard deviation away from the mean.<br>

# QQ Plot 
- Full form of QQ is Quantile-Quantile.
- The quantile-quantile (q-q) plot is a graphical technique for determining if two data sets come from populations with a common distribution.
- Q-Q plots are also used to find the Skewness (a measure of “asymmetry”) of a distribution. When we plot theoretical quantiles on the x-axis and the sample quantiles whose distribution we want to know on the y-axis then we see a very peculiar shape of a Normally distributed Q-Q plot for skewness. If the bottom end of the Q-Q plot deviates from the straight line but the upper end is not, then we can clearly say that the distribution has a longer tail to its left or simply it is left-skewed (or negatively skewed) but when we see the upper end of the Q-Q plot to deviate from the straight line and the lower and follows a straight line then the curve has a longer till to its right and it is right-skewed (or positively skewed).

```python
def QQ_plot(data, measure):
    fig = plt.figure(figsize = (20,7))
    (mu, sigma) = norm.fit(data)
    
    fig1 = fig.add_subplot(121)
    sns.distplot(data, fit = norm)
    fig1.set_title(measure + ' Distribution ( mean = {:.2f} and standard deviation = {:.2f} )'.format(mu, sigma), loc = 'center')
    fig1.set_xlabel(measure)
    fig1.set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

```python
QQ_plot(df.Mthly_HH_Income, 'Mthly_HH_Income')
```

# Box Cox
- A Box Cox transformation is a transformation of non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data isn’t normal, applying a Box-Cox means that you are able to run a broader number of tests.

- The Box Cox transformation is named after statisticians George Box and Sir David Roxbee Cox who collaborated on a 1964 paper and developed the technique.

```python
No_of_Earning_Members = df.No_of_Earning_Members
sns.distplot(No_of_Earning_Members)
G = probplot(No_of_Earning_Members , dist = 'norm' , plot = pylab)
pylab.show(G)
```
